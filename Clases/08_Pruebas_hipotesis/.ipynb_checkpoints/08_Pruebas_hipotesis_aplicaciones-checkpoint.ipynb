**Pruebas de hipótesis. Aplicaciones**

<font color = 'Black'>
<font size = 3>

Tenemos hasta ahora tres conceptos importantes asociados a los **contrastes de hipótesis**:

* **Estadístico de contraste**: es una **variable aleatoria** que va a tomar un valor dependiendo de la **muestra**, y observando ese valor vamos a tomar la decisión de si se acepta $H_{1}$ o se rechaza.

* **Nivel de significancia**.
* $p$-valor.

Para aceptar o rechazar se van a utilizar dos criterios:

1. el tamaño del $p$-valor.
2. el valor del **estadístico de contraste**.

De esta manera, si conocemos el nivel de significancia $\alpha$, la decisión que tomemos se puede basar en:

* **Región de rechazo:** si el estadístico de contraste cae dentro de una cierta región construída a partir de la significancia $\alpha$ (región crítica), rechazamos $H_0$.
* **El $p$-valor:** si el $p$-valor es mas pequeño que $\alpha$, rechazamos $H_0$. Si el $p$-valor es mas grande que $\alpha$, rechazamos $H_1$.

En este caso, conviene basar la decisión en ambos criterios.

En la siguiente imagen se muestra un esquema de la región crítica.

<img src="Img/region_critica.jpg">

Por otra parte, si no conocemos el nivel de significancia $\alpha$, la decisión que tomemos se basará en el $p$-valor: si el $p$-valor es pequeño ($<0.05$), rechazamos $H_0$; si el $p$-valor es grande ($>0.1$), rechazamos $H_1$; si el $p$-valor está entre 0.05 y 0.1, se recomienda realizar mas estudios.

**Pruebas de hipótesis en Python**

<font color = 'Black'>
<font size = 3>

En **Python**, no existen funciones predefinidas equivalentes directas a ``z.test`` (para medias), ``binom.test`` (para proporciones) y ``sigma.test`` (para desviaciones estándar) que se utilizan en **R**. Sin embargo, podemos implementar pruebas similares usando bibliotecas como ``scipy``, ``statsmodels`` y funciones básicas de Python. 

Parámetro poblacional|Desviación Poblacional|Muestra|Función
:--:|:--:|:--:|:--
media|conocida|gaussiana o grande| Algoritmo 1.
media|desconocida|gaussiana o grande|Algoritmo 2.
media|desconocida|pequeña|Algoritmo 3.
proporción|no aplica|no importa|Algoritmo 4.
varianza|no aplica|gaussiana o casi|Algoritmo 5.

En cada en cada una de ellas, el parámetro ``alternative`` se selecciona de la siguiente manera:

$H_1$|``alternative``
:--:|:--:
$\theta>\theta_0$|"greater"
$\theta<\theta_0$|"less"
$\theta\neq\theta_0$|"two-sided"

**Algoritmo 1.**

Código para resolver el problema **parámetro poblacional:** **media** - **desviación conocida** - **Muestra:** **gaussiana** o **grande**.

```python
'''
###############################################################################
###############################################################################
########### PRUEBA DE HIPÓTESIS PARA LA MEDIA CON VARIANZA CONOCIDA ###########
##########              POBLACION NORMAL O GRANDE                    ##########
###############################################################################
###############################################################################
'''

import numpy as np
from scipy.stats import norm

def z_test_mean(n, sample_mean, mu_0, sigma, alternative, alpha):
    """
    Prueba de hipótesis Z para la media poblacional con desviación estándar conocida.
    Supone que la población es normal o el tamaño de la muestra es suficientemente grande.

    Parameters:
    - n: tamaño de la muestra.
    - sample_mean: media muestral.
    - mu_0: media poblacional bajo la hipótesis nula (H0).
    - sigma: desviación estándar poblacional conocida.
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - z_stat: valor del estadístico Z.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Cálculo del estadístico Z
    z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        z_critical = norm.ppf(1 - alpha)
        rejection_region = (z_critical, np.inf)
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        z_critical = norm.ppf(alpha)
        rejection_region = (-np.inf, z_critical)
        p_value = norm.cdf(z_stat)
    elif alternative == 'two-sided':
        z_critical_low = norm.ppf(alpha / 2)
        z_critical_high = norm.ppf(1 - alpha / 2)
        rejection_region = ((-np.inf, z_critical_low), (z_critical_high, np.inf))
        p_value = 2 * min(norm.cdf(z_stat), 1 - norm.cdf(z_stat))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return z_stat, p_value, rejection_region, conclusion

#%%

n = 40
sample_mean = 725
sigma = 102
mu_0 = 670
alpha = 0.01
alternative='greater'

z_test_mean(n, sample_mean, mu_0, sigma, alternative, alpha)

#%%

sample = [2150, 1950, 2170, 1860, 2050, 2120, 1920, 1850, 2230]
n = len(sample)
sample_mean = np.mean(sample)
sigma = 210
mu_0 = 2000
alpha = 0.05
alternative='greater'

z_test_mean(n, sample_mean, mu_0, sigma, alternative, alpha)
```

**Algoritmo 2.**

Código para resolver el problema **parámetro poblacional:** **media**, **desviación desconocida**, **muestra gaussiana** o **grande**.

```python
'''
###############################################################################
###############################################################################
######### PRUEBA DE HIPÓTESIS PARA LA MEDIA CON VARIANZA DESCONOCIDA ##########
#########                       POBLACION CUALQUIER                  ##########
###############################################################################
###############################################################################
'''

import numpy as np
from scipy.stats import t

def t_test_mean(n, sample_mean, sample_std, mu_0, alternative, alpha):
    """
    Prueba de hipótesis t para la media poblacional con desviación estándar desconocida.
    Supone que la población es normal o el tamaño de la muestra es suficientemente grande.

    Parameters:
    - n: tamaño de la muestra.
    - sample_mean: media muestral.
    - sample_std: desviación estándar muestral.
    - mu_0: media poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - t_stat: valor del estadístico t.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Grados de libertad
    df = n - 1

    # Cálculo del estadístico t
    t_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        t_critical = t.ppf(1 - alpha, df)
        rejection_region = (t_critical, np.inf)
        p_value = 1 - t.cdf(t_stat, df)
    elif alternative == 'less':
        t_critical = t.ppf(alpha, df)
        rejection_region = (-np.inf, t_critical)
        p_value = t.cdf(t_stat, df)
    elif alternative == 'two-sided':
        t_critical_low = t.ppf(alpha / 2, df)
        t_critical_high = t.ppf(1 - alpha / 2, df)
        rejection_region = ((-np.inf, t_critical_low), (t_critical_high, np.inf))
        p_value = 2 * min(t.cdf(t_stat, df), 1 - t.cdf(t_stat, df))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return t_stat, p_value, rejection_region, conclusion

#%%

sample_std = 21
mu_0 = 880
alpha = 0.05
alternative = "two-sided"

#%%

sample_std = np.std(sample,ddof=1)
```

**Algoritmo 4.**

Código para resolver el problema con **parámetro poblacional:** **proporción**.

> ⚠️ **Nota importante:** **Aquí no importa el tamaño de la muestra.**

```python
'''
###############################################################################
###############################################################################
################    PRUEBA DE HIPÓTESIS PARA LA PROPORCIÓN    #################
###############################################################################
###############################################################################
'''

from scipy.stats import binom

def exact_test_proportion(n, x, p_0, alternative, alpha):
    """
    Prueba exacta para proporciones basada en la distribución binomial.

    Parameters:
    - n: tamaño de la muestra.
    - x: número de éxitos observados.
    - p_0: proporción poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - test_stat: valor del estadístico de prueba (número de éxitos observados).
    - p_value: valor p exacto de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # El estadístico de prueba es simplemente el número de éxitos observados
    test_stat = x

    # Cálculo del valor p según la hipótesis alternativa
    if alternative == 'greater':
        p_value = 1 - binom.cdf(x - 1, n, p_0)
        rejection_region = f"x >= {binom.ppf(1 - alpha, n, p_0):.0f}"
    elif alternative == 'less':
        p_value = binom.cdf(x, n, p_0)
        rejection_region = f"x <= {binom.ppf(alpha, n, p_0):.0f}"
    elif alternative == 'two-sided':
        # Calculamos ambos extremos para hipótesis bilateral
        p_low = binom.cdf(x, n, p_0)
        p_high = 1 - binom.cdf(x - 1, n, p_0)
        p_value = 2 * min(p_low, p_high)
        rejection_region = f"En ambos extremos, según p-value <= {alpha}"
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return test_stat, p_value, rejection_region, conclusion

#%%
n = 100
x = 15
p_0 = 0.2
alternative = "two-sided"
alpha = 0.05 
exact_test_proportion(n, x, p_0, alternative, alpha)
```

**Algoritmo 5.**

Código para resolver el problema con **parámetro poblacional:** **varianza** - **muestra:** **gaussiana** o **casi**.

```python
'''
###############################################################################
###############################################################################
#################    PRUEBA DE HIPÓTESIS PARA LA VARIANZA    ##################
###############################################################################
###############################################################################
'''

import numpy as np
from scipy.stats import chi2

def chi_squared_test_variance(n, sigma_squared, sigma_squared_0, alternative, alpha):
    """
    Prueba de hipótesis chi-cuadrado para la varianza.
    Supone que la población es normal.

    Parameters:
    - n: tamaño de la muestra.
    - s_squared: varianza muestral.
    - sigma_squared_0: varianza poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - chi2_stat: valor del estadístico chi-cuadrado.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Estadístico chi-cuadrado
    chi2_stat = (n - 1) * sigma_squared / sigma_squared_0

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        chi2_critical = chi2.ppf(1 - alpha, n-1)
        rejection_region = (chi2_critical, np.inf)
        p_value = 1 - chi2.cdf(chi2_stat, n-1)
    elif alternative == 'less':
        chi2_critical = chi2.ppf(alpha, n-1)
        rejection_region = (0, chi2_critical)
        p_value = chi2.cdf(chi2_stat, n-1)
    elif alternative == 'two-sided':
        chi2_critical_low = chi2.ppf(alpha / 2, n-1)
        chi2_critical_high = chi2.ppf(1 - alpha / 2, n-1)
        rejection_region = ((0, chi2_critical_low), (chi2_critical_high, np.inf))
        p_value = 2 * min(chi2.cdf(chi2_stat, n-1), 1 - chi2.cdf(chi2_stat, n-1))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return chi2_stat, p_value, rejection_region, conclusion

#%%

sample = [2.32, 4.26, 4.02, 4.44, 3.68, 2.72, 1.90, 1.21] 
n = len(sample)
s_squared = np.var(sample,ddof=1)
sigma_squared_0 = 0.8**2 
alpha = 0.05
alternative='two-sided'

chi_squared_test_variance(n, s_squared, sigma_squared_0, alternative, alpha)
```

**Ejemplo 1.**

<font color = 'LightSeaGreen'>
<font size = 3>

A un productor se le ofrece un nutriente con el que obtendrá rendimientos medios superiores a 2.000 kg/ha, que es la producción usual de la zona. Se sabe que en esa población los rendimientos tienen un **desvío estándar** de 210 kg/ha y que la distribución de los rendimientos es **gaussiana**.

El productor decide realizar una prueba sobre 9 parcelas, en las que pretende observar los rendimientos, promediarlos y de acuerdo al resultado optar o no por el nuevo producto: se acuerda correr un riesgo de rechazar el nutriente cuando en realidad éste cumple con las especificaciones promocionales, fijando para ello una probabilidad de error de 5%

Los resultados de las 9 parcelas fueron los siguientes: 2150, 1950, 2170, 1860, 2050, 2120, 1920, 1850, 2230.

¿Existe evidencia suficiente para decir que el nutriente aumenta los rendimientos?

**Solución.**

Lo que se quiere es mostrar que la **producción media** con tu producto es mayor a la usual. La usual es de 2000. Entonces, $\mu_0=2000$. Por lo que se busca probar que $\mu > \mu_{0}$.

Es claro que $\mu_0=2000$, $\sigma=210$, $\alpha=0.05$ (confianza al $95\%$). 

Ahora, recordemos la filosofía de las pruebas de hipótesis: se busca evidencia de que $H_1$ es verdadera. 

Por lo tanto, $H_1$ en este caso $H_1:\mu>2000$ (es decir, los rendimientos promedios se van a mejorar). 

De esta manera, se plantea el test:

$$\left\{\begin{array}{l}H_0:\mu\le2000\\H_1:\mu>2000\end{array}\right.$$ con $\sigma$ conocida e igual a 210 y $\alpha=0.05$, por lo que usamos el **Algoritmo 1.** para resolver el problema.

````python
'''
###############################################################################
###############################################################################
########### PRUEBA DE HIPÓTESIS PARA LA MEDIA CON VARIANZA CONOCIDA ###########
##########              POBLACION NORMAL O GRANDE                    ##########
###############################################################################
###############################################################################
'''

import numpy as np
from scipy.stats import norm

def z_test_mean(n, sample_mean, mu_0, sigma, alternative, alpha):
    """
    Prueba de hipótesis Z para la media poblacional con desviación estándar conocida.
    Supone que la población es normal o el tamaño de la muestra es suficientemente grande.

    Parameters:
    - n: tamaño de la muestra.
    - sample_mean: media muestral.
    - mu_0: media poblacional bajo la hipótesis nula (H0).
    - sigma: desviación estándar poblacional conocida.
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - z_stat: valor del estadístico Z.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Cálculo del estadístico Z
    z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        z_critical = norm.ppf(1 - alpha)
        rejection_region = (z_critical, np.inf)
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        z_critical = norm.ppf(alpha)
        rejection_region = (-np.inf, z_critical)
        p_value = norm.cdf(z_stat)
    elif alternative == 'two-sided':
        z_critical_low = norm.ppf(alpha / 2)
        z_critical_high = norm.ppf(1 - alpha / 2)
        rejection_region = ((-np.inf, z_critical_low), (z_critical_high, np.inf))
        p_value = 2 * min(norm.cdf(z_stat), 1 - norm.cdf(z_stat))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return z_stat, p_value, rejection_region, conclusion
````

````python
#%%

sample = [2150, 1950, 2170, 1860, 2050, 2120, 1920, 1850, 2230]
n = len(sample)
sample_mean = np.mean(sample)
sigma = 210
mu_0 = 2000
alpha = 0.05
alternative='greater'


est_cont, p_val, reg_rech, conclusion = z_test_mean(n, sample_mean, mu_0, sigma, alternative, alpha)

print(f"El estadístico de contraste es: {est_cont}")
print(f"El p-valor es: {p_val}")
print(f"La región de rechazo es: {reg_rech}")
print(f"Concluímos que: {conclusion}")
````

Con esto se concluye que no se rechaza $H_{0}$. 

No se está mostrando evidencia suficiente de que $H_{1}$ sea verdadera.

No convenciste al productor de que este producto es bueno.

Obervaciones:

- El estadístico de contraste no cayó dentro del intervalo. Si no cae dentro del intervalo no puedo rechazar $H_{0}$.

Por criterio de **estadístico de contraste** rechazo $H_{1}$.

- Como el p-valor que se obtuvo fue de 0.3169, entonces p-valor $> \alpha$. Por lo tanto no puedo rechazar $H_{0}$.

Por tanto, con cualquiera de los dos criterios no podemos rechazar $H_{0}$, entonces rechazamos $H_{1}$.

No se pudo convencer al campesino de que el producto mejoraba el rendimiento con un $95\%$ de confianza.

Se desea contrastar con un nivel del 5% (o confianza del $95\%$) la hipótesis de que la **altura media** de los hombres de 18 o más años de un país es menor a 180 cm. 

Se desconoce la **desviación poblacional** de las alturas, y se toma una **muestra** de 15 hombres dando como resultado 168, 168, 169, 167, 171, 172, 182, 175, 177, 173, 168, 195, 167, 175, y 175 (cm). 

¿Existe evidencia de que la altura media de los hombres mayores de edad de este país es menor a 180?

**Solución.**

Queremos estimar la **media** de las alturas, desconocemos la **desviación**, y la **muestra** fue de 15 hombres. Por lo que usaremos el **Algoritmo 3.** para resolver el problema.

Es claro que $\mu_0=180$ y $\alpha=0.05$ y lo que se quiere mostrar es que la estatura **promedio** es $< 180$ cm. 

Ahora, recordemos la filosofía de las pruebas de hipótesis: se busca evidencia de que $H_1$ es verdadera. 

Por lo tanto, $H_1$ en este caso $H_1:\mu<180$ (es lo que se quiere demostrar). 

De esta manera, se plantea el test:

$$\left\{\begin{array}{l}H_0:\mu\ge180\\H_1:\mu<180\end{array}\right.$$ con $\sigma$ **desconocida** y $\alpha=0.05$:

````python
import numpy as np
from scipy.stats import t

def t_test_mean(n, sample_mean, sample_std, mu_0, alternative, alpha):
    """
    Prueba de hipótesis t para la media poblacional con desviación estándar desconocida.
    Supone que la población es normal o el tamaño de la muestra es suficientemente grande.

    Parameters:
    - n: tamaño de la muestra.
    - sample_mean: media muestral.
    - sample_std: desviación estándar muestral.
    - mu_0: media poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - t_stat: valor del estadístico t.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Grados de libertad
    df = n - 1

    # Cálculo del estadístico t
    t_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        t_critical = t.ppf(1 - alpha, df)
        rejection_region = (t_critical, np.inf)
        p_value = 1 - t.cdf(t_stat, df)
    elif alternative == 'less':
        t_critical = t.ppf(alpha, df)
        rejection_region = (-np.inf, t_critical)
        p_value = t.cdf(t_stat, df)
    elif alternative == 'two-sided':
        t_critical_low = t.ppf(alpha / 2, df)
        t_critical_high = t.ppf(1 - alpha / 2, df)
        rejection_region = ((-np.inf, t_critical_low), (t_critical_high, np.inf))
        p_value = 2 * min(t.cdf(t_stat, df), 1 - t.cdf(t_stat, df))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return t_stat, p_value, rejection_region, conclusion
````

**Ejemplo 3.**

<font color = 'LightSeaGreen'>
<font size = 3>

En un estudio se afirma al menos 3 de cada 10 estudiantes universitarios trabajan. Pruebe esta aseveración, a un **nivel de significación** de 0.025, si una **muestra** aleatoria de 600 estudiantes universitarios revela que 200 de ellos trabajan.

**Solución.**

Lo que se quiere probar es que al menos 3 de cada 10 estudiantes universitarios trabajan. Es decir, si $p$ es la **proporción** verdadera de estudiantes que trabajan, entonces:

$H_1:p>0.3$.

Es claro que:

- $p_0 = \frac{3}{10} = 0.3$ es contra quién quiero comparar. 
- $n = 600$ es el número de personas que entrevisté.
- $x = 200$ es el número de personas que cumplen con la condición (trabajan).
- $\alpha = 0.025$.

Ahora, recordemos la filosofía de las pruebas de hipótesis: se busca evidencia de que $H_1$ es verdadera. Por lo tanto, $H_1$ en este caso es $H_1:p>0.3$ (es decir, al menos el 30% de los estudiantes universitarios trabajan). De esta manera, se plantea el test:

$$\left\{\begin{array}{l}H_0:p\le0.3\\H_1:p>0.3\end{array}\right.$$ con $x=200$, $n=600$ y $\alpha=0.025$

Para resolverlo usaremos el **Algoritmo 4.**

````python
from scipy.stats import binom

def exact_test_proportion(n, x, p_0, alternative, alpha):
    """
    Prueba exacta para proporciones basada en la distribución binomial.

    Parameters:
    - n: tamaño de la muestra.
    - x: número de éxitos observados.
    - p_0: proporción poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - test_stat: valor del estadístico de prueba (número de éxitos observados).
    - p_value: valor p exacto de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # El estadístico de prueba es simplemente el número de éxitos observados
    test_stat = x

    # Cálculo del valor p según la hipótesis alternativa
    if alternative == 'greater':
        p_value = 1 - binom.cdf(x - 1, n, p_0)
        rejection_region = f"x >= {binom.ppf(1 - alpha, n, p_0):.0f}"
    elif alternative == 'less':
        p_value = binom.cdf(x, n, p_0)
        rejection_region = f"x <= {binom.ppf(alpha, n, p_0):.0f}"
    elif alternative == 'two-sided':
        # Calculamos ambos extremos para hipótesis bilateral
        p_low = binom.cdf(x, n, p_0)
        p_high = 1 - binom.cdf(x - 1, n, p_0)
        p_value = 2 * min(p_low, p_high)
        rejection_region = f"En ambos extremos, según p-value <= {alpha}"
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return test_stat, p_value, rejection_region, conclusion
````

````python
#%%
n = 600 # número de personas que entrevisté
x = 200 # número de personas que trabajan
p_0 = 0.3 # contra qué voy a comparar
alternative = "greater"
alpha = 0.025

est_cont, p_val, reg_rech, conclusion = exact_test_proportion(n, x, p_0, alternative, alpha)

print(f"El estadístico de contraste es: {est_cont}")
print(f"El p-valor es: {p_val}")
print(f"La región de rechazo es: {reg_rech}")
print(f"Concluímos que: {conclusion}")
````

<font size = 3>

Con lo anterior, como el **estadístico de contraste** no está en el **intervalo**, no puedo rechazar $H_{0}$.
Como no podemos rechazar $H_{0}$, no podemos aceptar $H_{1}$ o que 3 de cada 10 estudiantes trabajan.
Con una confianza del $97.5\%$ no puedo aceptar que sea verdadero eso de que 3 de cada 10 estudiantes trabajan.

**Ejemplo 4.**

<font color = 'LightSeaGreen'>
<font size = 3>

Un fabricante de baterías garantiza que su producto dura en **promedio** 2.5 años con una **desviación estándar** de al menos 0.8 años (es decir, $\sigma>0.8$). Se sabe que la vida de uso de estas baterías sigue una **distribución normal**.

Se toma una **muestra** aleatoria de 8 baterías y resultó que la duración de cada una fue 2.32, 4.26, 4.02, 4.44, 3.68, 2.72, 1.90, 1.21 (años).

Con la evidencia tomada, ¿tiene razón el fabricante respecto a la **desviación estándar poblacional**? Utilice un 95% de confianza.

**Solución.**

El fabricante dice que sus baterías duran entre 1.7 y 2.3. Es decir, $2.5 \pm 0.8$ años.

En la siguiente imagen se muestra un esquema de la vida de las baterías que promete el fabricante.

<img src="Img/lifetime.jpg">

Lo que se quiere probar es que la $\sigma > 0.8$, pero la prueba que tenemos es para las **varianzas** no para las **desviaciones**.

Recordar que **desviación estándar** $ \sigma = \sqrt(var)$. Entonces, para aplicar la prueba haremos $ \sigma^{2} = var$.

es la **desviación al cuadrado**.

Es claro que $\sigma_0=0.8$ y $\alpha=0.05$. Ahora, recordemos la filosofía de las pruebas de hipótesis: se busca evidencia de que $H_1$ es verdadera. 

Por lo tanto, $H_1$ en este caso es $H_1:\sigma^{2} > (0.8)^{2}$ (es decir, la **desviación** de la duración de la batería en años no es 0.8). 

De esta manera, se plantea el test:

$$\left\{\begin{array}{l}H_0:\sigma^{2} = (0.8)^{2}\\H_1:\sigma^{2} > (0.8)^{2}\end{array}\right.$$ con $\alpha=0.05$:

Por lo que usaremos el **Algoritmo 5.** para resolverlo.

````python
import numpy as np
from scipy.stats import chi2

def chi_squared_test_variance(n, sigma_squared, sigma_squared_0, alternative, alpha):
    """
    Prueba de hipótesis chi-cuadrado para la varianza.
    Supone que la población es normal.

    Parameters:
    - n: tamaño de la muestra.
    - s_squared: varianza muestral.
    - sigma_squared_0: varianza poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - chi2_stat: valor del estadístico chi-cuadrado.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Estadístico chi-cuadrado
    chi2_stat = (n - 1) * sigma_squared / sigma_squared_0

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        chi2_critical = chi2.ppf(1 - alpha, n-1)
        rejection_region = (chi2_critical, np.inf)
        p_value = 1 - chi2.cdf(chi2_stat, n-1)
    elif alternative == 'less':
        chi2_critical = chi2.ppf(alpha, n-1)
        rejection_region = (0, chi2_critical)
        p_value = chi2.cdf(chi2_stat, n-1)
    elif alternative == 'two-sided':
        chi2_critical_low = chi2.ppf(alpha / 2, n-1)
        chi2_critical_high = chi2.ppf(1 - alpha / 2, n-1)
        rejection_region = ((0, chi2_critical_low), (chi2_critical_high, np.inf))
        p_value = 2 * min(chi2.cdf(chi2_stat, n-1), 1 - chi2.cdf(chi2_stat, n-1))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return chi2_stat, p_value, rejection_region, conclusion
````

````python
#%%

muestra = [2.32, 4.26, 4.02, 4.44, 3.68, 2.72, 1.90, 1.21]
n = len(muestra)
sigma_squared = np.var(muestra, ddof = 1) # sigma cuadrada es la varianza muestral
sigma_squared_0 = 0.8**2 
alpha = 0.05
alternative='greater'

est_cont, p_val, reg_rech, conclusion = chi_squared_test_variance(n, sigma_squared, sigma_squared_0, alternative, alpha)

print(f"El estadístico de contraste es: {est_cont}")
print(f"El p-valor es: {p_val}")
print(f"La región de rechazo es: {reg_rech}")
print(f"Concluímos que: {conclusion}")
````

<font color = 'LightSeaGreen'>
<font size = 3>

Con lo anterior, como el **estadístico de contraste** no está en el **intervalo**, no puedo rechazar $H_{0}$.
Como no podemos rechazar $H_{0}$, no podemos aceptar $H_{1}$ o que 3 de cada 10 estudiantes trabajan.
Con una confianza del $97.5\%$ no puedo aceptar que sea verdadero eso de que 3 de cada 10 estudiantes trabajan.

Como el **estadístico de contraste** si cae dentro del **intervalo**, por el **criterio de región** rechazamos $H_{0}$.

Como el **p-valor** es **más pequeño** que $\alpha$, $0.02748 < 0.05$. Por lo tanto, con el criterio del **p-valor**, rechazamos rechazamos $H_{0}$.

Concluimos que rechazamos $H_{0}$. Como rechazar $H_{0}$ implica aceptar $H_{1}$. Por lo tanto, tenemos evidencia suficiente para decir que elo fabricante dice la verdad.

Las etiquetas en latas de 1 galón de pintura por lo general indican el tiempo de secado y el area que puede cubrir una capa. Casi todas las marcas indican que en una capa, un galón cubrirá entre 250 y 500 ft$^2$. 

Un fabricante afirma que un galón de su pintura cubrirá 400ft$^2$. 

Para probar su afirmación se toma una muestra aleatoria de 10 pinturas. 

Las áreas cubiertas por cada pintura, en ft$^2$, fueron 310, 311, 412, 368, 447, 376, 303, 410, 365, 350.

Una lata de 1 galón de pintura usualmente cubre entre entre 250 y 500 ft$^2$. Pero llega otro fabricante y dice que la que él produce, en promedio es siempre es de 400ft$^2$.

````python
import numpy as np
from scipy.stats import t

def t_test_mean(n, sample_mean, sample_std, mu_0, alternative, alpha):
    """
    Prueba de hipótesis t para la media poblacional con desviación estándar desconocida.
    Supone que la población es normal o el tamaño de la muestra es suficientemente grande.

    Parameters:
    - n: tamaño de la muestra.
    - sample_mean: media muestral.
    - sample_std: desviación estándar muestral.
    - mu_0: media poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - t_stat: valor del estadístico t.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Grados de libertad
    df = n - 1

    # Cálculo del estadístico t
    t_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        t_critical = t.ppf(1 - alpha, df)
        rejection_region = (t_critical, np.inf)
        p_value = 1 - t.cdf(t_stat, df)
    elif alternative == 'less':
        t_critical = t.ppf(alpha, df)
        rejection_region = (-np.inf, t_critical)
        p_value = t.cdf(t_stat, df)
    elif alternative == 'two-sided':
        t_critical_low = t.ppf(alpha / 2, df)
        t_critical_high = t.ppf(1 - alpha / 2, df)
        rejection_region = ((-np.inf, t_critical_low), (t_critical_high, np.inf))
        p_value = 2 * min(t.cdf(t_stat, df), 1 - t.cdf(t_stat, df))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return t_stat, p_value, rejection_region, conclusion
````

````python
muestra = [310, 311, 412, 368, 447, 376, 303, 410, 365, 350]
n = len(muestra)
sample_mean = np.mean(muestra)
sample_std = np.std(muestra, ddof = 1)
mu_0 = 400
alternative = "two-sided"
alpha = 0.05

est_cont, p_value, reg_rech, conclusion = t_test_mean(n, sample_mean, sample_std, mu_0, alternative, alpha)

print(f"El estadístico de contraste es: {est_cont}")
print(f"El p-valor es: {p_val}")
print(f"La región de rechazo es: {reg_rech}")
print(f"Concluímos que: {conclusion}")
````

**Ejemplo 6.**

La producción diaria de una planta química ha **promediado** 880 ton en los últimos años. A la gerente de control de calidad le gustaría saber si este promedio ha cambiado en meses recientes. 

Selecciona 50 días de las base de datos y calcula el **promedio** y la **desviación**. Obtiene que el **promedio** es de 871 ton con una **desviación muestral** de 21. Pruebe la **hipótesis** apropiada usando $\alpha = 0.05$

**Solución.** 

Lo que queremos probar es si sigue produciendo 880 ton.

Nos piden testear $H_1:\, \mu \neq  880$.

Para este problema usamos el **Algoritmo 2.**

````python
import numpy as np
from scipy.stats import t

def t_test_mean(n, sample_mean, sample_std, mu_0, alternative, alpha):
    """
    Prueba de hipótesis t para la media poblacional con desviación estándar desconocida.
    Supone que la población es normal o el tamaño de la muestra es suficientemente grande.

    Parameters:
    - n: tamaño de la muestra.
    - sample_mean: media muestral.
    - sample_std: desviación estándar muestral.
    - mu_0: media poblacional bajo la hipótesis nula (H0).
    - alternative: tipo de hipótesis ('two-sided', 'greater', 'less').
    - alpha: nivel de significancia (default: 0.05).

    Returns:
    - t_stat: valor del estadístico t.
    - p_value: valor p de la prueba.
    - rejection_region: región de rechazo de H_0.
    - conclusion: decisión sobre H_0 (rechazo o no rechazo).
    """
    # Grados de libertad
    df = n - 1

    # Cálculo del estadístico t
    t_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

    # Cálculo de la región de rechazo y valor p según la hipótesis alternativa
    if alternative == 'greater':
        t_critical = t.ppf(1 - alpha, df)
        rejection_region = (t_critical, np.inf)
        p_value = 1 - t.cdf(t_stat, df)
    elif alternative == 'less':
        t_critical = t.ppf(alpha, df)
        rejection_region = (-np.inf, t_critical)
        p_value = t.cdf(t_stat, df)
    elif alternative == 'two-sided':
        t_critical_low = t.ppf(alpha / 2, df)
        t_critical_high = t.ppf(1 - alpha / 2, df)
        rejection_region = ((-np.inf, t_critical_low), (t_critical_high, np.inf))
        p_value = 2 * min(t.cdf(t_stat, df), 1 - t.cdf(t_stat, df))
    else:
        raise ValueError("El parámetro 'alternative' debe ser 'two-sided', 'greater' o 'less'.")

    # Decisión basada en el valor p
    if p_value < alpha:
        conclusion = f"Rechazamos H0 al nivel de significancia {alpha}."
    else:
        conclusion = f"No rechazamos H0 al nivel de significancia {alpha}."

    return t_stat, p_value, rejection_region, conclusion
````

````python
n = 50
sample_mean = 871
sample_std = 21
alpha = 0.05
mu_0 = 880
alternative = "two-sided"

est_cont, p_val, reg_rech, conclusion = t_test_mean(n, sample_mean, sample_std, mu_0, alternative,alpha)

print(f"El estadístico de contraste es {est_cont}")
print(f"El p-valor es {p_val}")
print(f"La región de rechazo es {reg_rech}")
print(f"Concluímos que {conclusion}")
````

<font color = 'LightSeaGreen'>
<font size = 3> 

Como el **estadístico de contraste** cae dentro del intervalo, entonces rechazamos $H_{0}$, por lo que aceptamos $H_1:\,\mu \neq 880$.

Con esto se concluye que si ha cambiado la cantidad de producción.

**Notas**

<font color = 'DarkRed'>
<font size = 3>

- Siempre se busca aceptar $H_{1}$, esa es la tirada, siempre demostrar la **hipótesis alternativa**.
- Siempre se trabaja sobre la **hipótesis alternativa**.
> ⚠️ **Nota importante: Rechazar $H_{1}$ (hipótesis alternativa) no significa que $H_{0}$ sea verdad**.
- Rechazar $H_{1}$ significa que no tienes evidencia que valide tu hipótesis.
> ⚠️ **Nota importante: Rechazar $H_{1}$ (hipótesis alternativa) no significa aceptar $H_{0}$**.
- Usualmente el valor de $\alpha$ lo propones tú.
- Cuando la **población** es pequeña y la **desviación** es **desconocida** ahí lo que te puede ayudar es que la **distribución** del histograma tenga forma de **montaña**. Si tiene forma de **montaña** ya la hiciste.
- Si no tiene forma de **montaña** el histograma, no puedes tener tanta **confianza** en los resultados.
- $S$ es la **desviación muestral** y se comporta como la $\chi^2$ con $n - 1$ grados de libertad.
- El p-valor siempre es el mismo sin importar en valor de $\alpha$.
- Las **regiones de rechazo** siempre dependen de la $\alpha$.
> ⚠️ **Nota importante: Siempre que se rechaza $H_{0}$ se acepta $H_{1}$**.
- Las **pruebas de hipótesis** son importantes en la industria, ya que de aquí arrancas para hacer pruebas de calidad.
- Existen dos pruebas para $\chi^2$:
  1. Para datos categóricos para medir la frecuencia.
  2. Para **pruebas de hipótesis** sobre la **desviación**.

**Deducción de la prueba de hipótesis unilateral para la media poblacional de una población gaussiana  con desviación conocida**

<font color = 'Black'>
<font size = 3>

Sea $X$ una **característica poblacional** con **distribución gaussiana**, donde se **desconoce la media** $\mu$ pero se **conoce la desviación** $\sigma$. 

Fijamos $\mu_0$ y consideremos el **contraste** 

$$\left\{\begin{array}{l}H_0:\mu=\mu_0\\H_1:\mu>\mu_0\end{array}\right.$$

con una **significancia** $\alpha$.

Tomemos una **muestra** aleatoria $X_1, X_2,...,X_n$ y definamos $Z$ como $Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}$. 

Nota que si $H_0$ es verdadera, entonces $Z$ es una **gaussiana** con **media** 0 y **desviación** 1.

Recordemos que buscamos evidencia de que $H_0$ no es verdadera, por lo cual si $\overline{X}$, la **media muestral**, es "bastante" más grande que $\mu_0$, tendremos esa evidencia. 

En símbolos: si $\overline{X}>A$ para un cierto valor $A$ apropiado, rechazaremos $H_0$. Como $Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt(n)}$, esto equivale a $Z>A^\prime$ para alguna $A^\prime$. 

Ahora bien, la definición de $\alpha$ nos dice $\alpha=P(\mbox{rechazar }H_0\mbox{ siendo verdadera})$, de donde $\alpha=P(Z>A^\prime)$, así que $1-\alpha=P(Z\le A^\prime)$, de modo que $A^\prime$ es el cuantil $1-\alpha$ de una **normal estándar**: $z_{1-\alpha}$ (lo que en **R** se calcula como $A^\prime$=``qnorm(1-alpha,1,0)``)

De esta manera: **Rechazamos $H_0$ si $Z>z_{1-\alpha}$**. Pero como $Z=\frac{\overline{X}-\mu_0}{\sigma/\sqrt{n}}$, lo anterior equivale a $$\overline{X}-z_{1-\alpha}\frac{\sigma}{\sqrt{n}}>\mu_0$$

Esto es: **Rechazamos $H_0$ si** $$\mu_0\in\left(\overline{X}-z_{1-\alpha}\frac{\sigma}{\sqrt{n}},\infty\right)$$
